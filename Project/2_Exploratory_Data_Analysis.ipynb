```python
import pandas as pd

# Load the data
df = pd.read_csv(r"C:\Users\dasc5\Downloads\E-commerce Furniture Dataset 2024\Data\ecommerce_furniture_dataset_2024.csv")
```

```python
# Distribution of Units Sold
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))  # Set appropriate figure size
sns.histplot(df['sold'], kde=True, color='skyblue', bins=30)

plt.title('Distribution of Units Sold')
plt.xlabel('Units Sold')
plt.ylabel('Count')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Units Sold by Price Range
df['price_numeric'] = df['price'].replace('[\$,]', '', regex=True).astype(float)

df['price_bin'] = pd.cut(df['price_numeric'], bins=[0, 50, 100, 200, 1000], 
                         labels=['Low', 'Medium', 'High', 'Premium'])

sns.boxplot(x='price_bin', y='sold', data=df)
plt.title('Units Sold by Price Range')
plt.xlabel('Price Range')
plt.ylabel('Units Sold')
```

```python
# Price vs Units Sold
sns.scatterplot(x='price_numeric', y='sold', data=df)
plt.title('Price vs. Units Sold')
plt.xlabel('Price')
plt.ylabel('Units Sold')
plt.show()
```

```python
# Units Sold by Shipping Type
df['tagText'] = df['tagText'].apply(lambda x: x if x in ['Free shipping', '+Shipping: $5.09'] else 'others')

# Then plot
plt.figure(figsize=(8, 5))
sns.boxplot(x='tagText', y='sold', data=df)
plt.title('Units Sold by Shipping Type')
plt.ylabel('Units Sold')
plt.xlabel('Shipping Type')
plt.grid(True)
plt.show()
```
